```
import keras
```

```
model=keras.Sequential([keras.layers.Dense(1,input_shape=[1])])  #sequential is used because it is stack of layers in sequence
```

```
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```
import numpy as np
```

```
xs=np.array([-2,-1,0,1,2,3,4],dtype=float)
```

```
ys=np.array([-5,-3,-1,1,3,5,7],dtype=float)
```

```
#training takes place in fit command
```

```
model.fit(xs,ys,epochs=500)
```

```
model.predict([10])
```
