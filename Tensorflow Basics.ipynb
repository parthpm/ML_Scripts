```
const=tf.constant(10)
```

```
fill_mat=tf.fill((4,4,),10)
```

```
zeros=tf.zeros((4,4))
```

```
ones=tf.ones((3,3,))
```

```
randn=tf.random_normal((4,4),mean=2,stddev=1)
```

```
randu=tf.random_uniform((4,4),minval=1,maxval=2)
```

```
randn
```

```
operation_list=[const,fill_mat,zeros,ones,randn,randu]
```

```
#ALterate to with tf.Session() as sess:
sess=tf.InteractiveSession()
```

```
for op in operation_list:
  print(sess.run(op)) #or print(op.eval())
  print("\n")
```
